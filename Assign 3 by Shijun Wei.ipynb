```python
from bs4 import BeautifulSoup
import requests
import time

# (a)
def main():
    try:
        headers = {'User-agent': 'Mozilla/5.0'} 

        URL = "https://www.ebay.com/sch/i.html?_nkw=iphone+6s"
        page = requests.get(URL, headers=headers)
                
    except Exception as ex:
        print("Error:" + str(ex))
        
    with open('ebay_iphone6s_01.htm','w', encoding='utf-8') as file:
        file.write(page.text)
        
if __name__ == '__main__':
    main()
 
```

```python
 #(b)
for x in range(1,11):
    headers = {'User-agent': 'Mozilla/5.0'} 
    URL = "https://www.ebay.com/sch/i.html?_nkw=iphone+6s&_pgn="+str(x)
    page = requests.get(URL, headers=headers)
    time.sleep(10) 
    filename = 'ebay_iphone6s_'+ str(x) +'.htm'
    time.sleep(10) 
    with open(filename,'w', encoding='utf-8') as file:
        file.write(page.text)
```

```python
#(c)
for x in range(1,11):
    filename = 'ebay_iphone6s_'+ str(x) +'.htm'
    with open(filename, 'r',encoding='utf-8') as file:
        text = file.read()
        soup = BeautifulSoup(text, 'html.parser')
        div = soup.find_all("div", class_= "s-item__info clearfix")
        for i in div:
            if(i.find('div',style="display:flex;")) :
                links = i.select('a', class_= 's-item__link')
                url = links[0].get('href')
                print(url)
```
